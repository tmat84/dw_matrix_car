```
def ploting_bar(feat_groupby,feat_agg='price_value',
                agg_func=[np.mean,np.median,np.size],feat_sort='mean'):
  return (
      data
      .groupby(feat_groupby)[feat_agg]
      .agg(agg_func)
      .sort_values(by=feat_sort,ascending=False)
      .head(50)

  ).plot(kind='bar',figsize=(15,5),subplots=True)
```

```
ploting_bar('param_marka-pojazdu');
```

```
ploting_bar('param_kraj-pochodzenia',feat_sort='size');
```
